```python
x = ds.iloc[:, [3, 4]].values
x
```

# Use dendogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.ylabel('Euclidean Distance')
plt.xlabel('Customers')
plt.show()
```

# Training the Hierarchial Clustering for the given dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
y_hc
```

```python
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 200, c = 'red', label = 'cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 200, c = 'green', label = 'cluster 1')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 200, c = 'blue', label = 'cluster 1')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 200, c = 'yellow', label = 'cluster 1')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 200, c = 'cyan', label = 'cluster 1')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.title('clusters of customers')
plt.legend()
plt.show()
```
